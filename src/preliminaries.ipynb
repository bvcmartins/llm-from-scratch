```python
import math 
import matplotlib.pyplot as plt

def sigmoid(x):
    """Compute the sigmoid function."""
    return 1 / (1 + math.exp(-x))
```

```python
def log_sigmoid(x):
    """Compute the log of the sigmoid function."""
    return math.log(sigmoid(x))
```

The log-sigmoid function is used as the loss for PPO. 

Discussion here: https://magazine.sebastianraschka.com/p/llm-training-rlhf-and-its-alternatives

```python
sigmoid_values = []
log_sigmoid_values = []
x_values = list(range(-10, 11))
for i in x_values:
    sigmoid_values.append(sigmoid(i))
    log_sigmoid_values.append(log_sigmoid(i))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x_values, sigmoid_values, label='Sigmoid', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid()
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_values, log_sigmoid_values, label='Log Sigmoid', color='red')
plt.title('Log Sigmoid Function')
plt.xlabel('x')
plt.ylabel('log(sigmoid(x))')
plt.grid()
plt.legend()
```
